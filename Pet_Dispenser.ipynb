# Descrição da Ideia



Feedo Smart é uma solução inovadora que controla e analisa a venda de rações para pets nas lojas. Ele usa sensores para ajustar a quantidade de ração e coletar dados precisos sobre o estoque, o consumo e a preferência dos clientes. Com o Feedo Smart, você pode oferecer uma alimentação personalizada e saudável para os pets dos seus clientes, além de aumentar a lucratividade e fidelização.

**Arquitetura Inicial do Projeto**

Funcionalidades:

*   Escolha da quantidade de ração
*   Porção de forma automática disponibilizada para o cliente.

Arquitetura:



*   Esp 32 (Placa responsavel pela conexão com a cloud e processamento de informações.)

*   Servo Motor (Responsável pela abertura do dispenser para liberar a ração.)

*   Célula de carga (Esta irá medir o peso da ração liberada para assim informar se precisa de mais ou não.)

*   Display (Retorna para o cliente o peso que foi fornecido e se está concluido já.)

*   Cloud (Onde será armazenado informaçoes de consumo, estoque e etc)

*   Website (Visualizaçao dos dados)
